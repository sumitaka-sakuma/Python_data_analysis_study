```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)
dates = pd.date_range(start="2020/04/01", periods=365)
dates
```

```python
df = pd.DataFrame(np.random.randint(1, 31, 365), index=dates, columns=["rand"])
df
```

```python
df_year = pd.DataFrame(df.groupby(pd.Grouper(freq='W-MON')).sum(), columns=["rand"])
df_year
```
