```python
import matplotlib.pyplot as plt
import matplotlib.style
```

```python
# ggplotスタイルを指定
matplotlib.style.use('ggplot')
```

```python
# データを用意
x = [1,2,3]
y = [2,4,9]

# 折れ線グラフを描画
plt.plot(x,y)
plt.title('MATLAB-style')

plt.show()
```

```python
# データを用意
x = [1,2,3]
y = [2,4,9]

# 描画オブジェクト(fig)とサブプロット（ax）を生成
fig, ax = plt.subplots()

# 折れ線グラフを描画
ax.plot(x, y)
ax.set_title('OOP-style')

plt.show()
```

```python
# 2つのサブプロットを配置
fig, axes = plt.subplots(2)
plt.show
```

```python
# ２行2列のサブプロットを配置
fig, axes = plt.subplots(2, 2)
plt.show()
```

```python
# 1行2列のサブプロットを配置
fig, axes = plt.subplots(ncols=2)
plt.show()
```
