```python
import matplotlib.pyplot as plt
import matplotlib.style
```

```python
# ggplotスタイルを指定
matplotlib.style.use('ggplot')
```

```python
# データを用意
x = [1,2,3]
y = [2,4,9]

# 折れ線グラフを描画
plt.plot(x,y)
plt.title('MATLAB-style')

plt.show()
```

```python
# データを用意
x = [1,2,3]
y = [2,4,9]

# 描画オブジェクト(fig)とサブプロット（ax）を生成
fig, ax = plt.subplots()

# 折れ線グラフを描画
ax.plot(x, y)
ax.set_title('OOP-style')

plt.show()
```

```python
# 2つのサブプロットを配置
fig, axes = plt.subplots(2)
plt.show
```

```python
# ２行2列のサブプロットを配置
fig, axes = plt.subplots(2, 2)
plt.show()
```

```python
# 1行2列のサブプロットを配置
fig, axes = plt.subplots(ncols=2)
plt.show()
```

```python
# スタイルの一覧を取得
print(matplotlib.style.available)
```

```python
# グラフのスタイルにclassicを指定
matplotlib.style.use('classic')

fig, ax = plt.subplots()
ax.plot([1, 2])

plt.show()
```

```python
fig, axes = plt.subplots(ncols=2)

# サブプロットにタイトルを設定
axes[0].set_title('subplot title 0')
axes[1].set_title('subplot title 1')
# 描画オブジェクトにタイトルを設定
fig.suptitle('figure title')

plt.show()
```

```python
fig, ax = plt.subplots()

# ラベルを設定
ax.set_xlabel('x label')
ax.set_ylabel('y label')

plt.show()
```

```python
fig, ax = plt.subplots()

# 凡例用のラベルを設定
ax.plot([1,2,3], [2,4,9], label='legend label')
ax.legend(loc='best')

plt.show()
```

```python
fig, ax = plt.subplots()

# 凡例用のラベルを設定
ax.plot([1,2,3], [2,4,9], label='legend label')
# 凡例を右下に表示
ax.legend(loc='lower right')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_title('subplot title')

#保存
fig.savefig('sample_figure.png')
fig.savefig('sample_figure.svg')
```
