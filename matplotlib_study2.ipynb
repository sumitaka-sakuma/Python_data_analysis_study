```python
import matplotlib.pyplot as plt
import matplotlib.style
```

```python
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [1,2,3]
y2 = [3,1,2]
ax.plot(x, y1)
ax.plot(x, y2)

plt.show()
```

```python
import numpy as np

x = np.arange(0.0, 15.0, 0.1)

y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()

plt.show()
```
