```python
import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
```

```python
fig, ax = plt.subplots()

# 線の色を名前で指定
ax.plot([1, 3], [3, 1], label='aqua', color='aqua')
# １６進数のRGBで指定
ax.plot([1, 3], [1, 3], label='#0000FF', color='#0000FF')
# RGBAをfloatで指定
ax.plot([1, 3], [2, 2], label='(0.1, 0.2, 0.5, 0.3)',
       color=(0.1, 0.2, 0.5, 0.3))
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

# 塗りつぶしの色を指定
ax.bar([1], [3], color='aqua')
# 塗りつぶしの色と枠線の色を指定
ax.bar([2], [4], color='aqua', edgecolor='black')

plt.show()
```
