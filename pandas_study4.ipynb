```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("20210401health.xlsx")
```

```python
df
```

```python
# 条件の抽出
df["歩数"] >= 10000
```

```python
# 歩数が１００００以上のものだけ抽出
df_selected = df[df["歩数"] >= 10000]
df_selected
```

```python
df_selected.shape
```

```python
# query
df.query('歩数 >= 10000 and 摂取カロリー <= 2000')
```

```python
df.dtypes
```

```python
# 日付をobjectからdatetimeに変換
df.loc[:, 'date'] = df.loc[:, '日付'].apply(pd.to_datetime)
```

```python
df.loc[:, "date"]
```

```python
df
```

```python
# 摂取カロリーを小数に変換
df.loc[:, "摂取カロリー"] = df.loc[:, "摂取カロリー"].astype(np.float32)
```

```python
df = df.set_index("date")
```

```python
df.head()
```

```python
# 並び替え
df.sort_values(by="歩数")
```

```python
# 降順
df.sort_values(by="歩数", ascending=False)
```

```python
# 歩数を摂取カロリーで割った値を、新しく作ったカラムに挿入
df.loc[:, "歩数/カロリー"] = df.loc[:, "歩数"] / df.loc[:,"摂取カロリー"]
df
```

```python
# 歩数/カロリーを元に、運動指数を返却する関数を作成
def exercise_judge(ex):
    if ex <= 3.0:
        return "Low"
    elif 3.0 < ex <= 6.0:
        return "Mid"
    else:
        return "High"

# 関数の返却値を運動指数カラムに格納
df.loc[:, "運動指数"] = df.loc[:, "歩数/カロリー"].apply(exercise_judge)
df
```

```python
# DataFrameを保存
df.to_pickle("df_202104health.pickle")
```

```python
df_moved = pd.get_dummies(df.loc[:, "運動指数"], prefix = "運動")
df_moved
```

```python
# 保存
df_moved.to_pickle("df_202104moved.pickle")
```
