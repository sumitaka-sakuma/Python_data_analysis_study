```python
import numpy as np
import pandas as pd
```

```python
# 1ヶ月分の日付を取得
dates = pd.date_range(start = "2021/04/01", end = "2021/04/30")
dates
```

```python
# 乱数を設定
np.random.seed(123)
df = pd.DataFrame(np.random.randint(1, 31, 30), index=dates, columns=["乱数"])
df
```

```python
# 1年分のデータを作成
dates = pd.date_range(start="2021/01/01", periods=365)
dates
```

```python
dates.shape
```

```python
# 乱数を設定
np.random.seed(123)
df = pd.DataFrame(np.random.randint(1,31,365), index=dates, columns=["乱数"])
df
```

```python
# 月平均を求める
df.groupby(pd.Grouper(freq="M")).mean()
```
