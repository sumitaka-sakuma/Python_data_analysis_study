```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('202103health.xlsx',
                   index_col="日付", 
                   parse_dates=True)
df
```

```python
# 欠損値の行の削除
df_drop = df.dropna()
df_drop
```

```python
# 欠損値に０を格納
df_fillna = df.fillna(0)
df_fillna 
```

```python
# 欠損値を一つ手前の行の値で補完
df_fill = df.fillna(method='ffill')
df_fill
```

```python
# 欠損値を他の値の平均値で補完
df_fillmean = df.fillna(df.mean())
df_fillmean
```
