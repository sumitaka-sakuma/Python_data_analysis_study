```python
import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
import pandas as pd
```

```python
# スタイルを指定
matplotlib.style.use('ggplot')

# DataFrameを作成
df = pd.DataFrame({'A': [1,2,3], 'B': [3,1,2]})
df.plot()

plt.show()
```

```python
np.random.seed(123)
df2 = pd.DataFrame(np.random.rand(3,2),
                  columns=['y1', 'y2'])
df2.plot.bar()
plt.show()
```

```python
# 積み上げ棒グラフを描画
df2.plot.bar(stacked=True)
plt.show()
```
